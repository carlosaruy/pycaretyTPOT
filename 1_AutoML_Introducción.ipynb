## Proceso de Data Science 

Durante la etapa de modelado, Data Scientist est resolviendo la tarea de optimizacin para maximizar la precisin del modelo. El proceso es complicado y requiere diferentes tipos de habilidades:
1.	La ingeniera de atributos
2.	La optimizacin de los hiperparmetros requiere una comprensin profunda de los algoritmos y los conceptos bsicos de ML;
3.	Se necesitan habilidades de ingeniera de software para que el cdigo de salida sea fcil de comprender e implementar.


### Qu hace autoML

AutoML trata de encontrar convinaciones de modelos y de hisperparametros que maximicen la precision de los modelos.Para ello:

- AutoML elige una estrategia para preprocesar los datos: cmo tratar los datos desequilibrados, cmo llenar los valores faltantes, eliminar, reemplazar o mantener valores atpicos, cmo codificar categoras y columnas de categoras mltiples, cmo prevenir errores de memoria...
- AutoML genera nuevas funciones y selecciona las significativas
- AutoML selecciona el modelo (modelos lineales, K-vecinos ms cercanos, aumento de gradiente, redes neuronales...)
- AutoML ajusta los hiperparmetros del modelo elegido 
- AutoML crea un conjunto estable de modelos para aumentar la puntuacin si es posible

# TPOT

Es una herramienta de aprendizaje automtico automatizado de Python que optimiza los procesos de aprendizaje automtico mediante programacin gentica.

**[Click here to read the documentation](http://epistasislab.github.io/tpot/)**

Source: [Github](https://github.com/EpistasisLab/tpot)

TPOT automatizar la parte ms tediosa del aprendizaje automtico mediante la exploracin inteligente de miles de posibles conbinaciones para encontrar la mejor para sus datos.

<img src="https://raw.githubusercontent.com/EpistasisLab/tpot/master/images/tpot-ml-pipeline.png">

Una vez que TPOT ha terminado de buscar (o se cansa de esperar), le proporciona el cdigo Python para la mejor canalizacin que encontr para que pueda jugar con la canalizacin desde all.

<img src="https://raw.githubusercontent.com/EpistasisLab/tpot/master/images/tpot-pipeline-example.png">

**Es una biblioteca de alto nivel construida sobre numpy, pandas, tqdm, sklearn, DEAP, etc., por lo que es muy fcil de usar.**


Generar un modelo con TPOT es como si se entrenara utilizando el mejor modelo para nuestro caso de la libreria de scikit-learn, ya que se ha construido sobre l. Este modelo estar altamente optimizado y brindar los mejores resultados. 

Cada vez que ejecutemos el kernel obtendremos diferentes modelos para los mismos datos porque cada vez que ejecuta el kernel, debido al enfoque de programacin gentica de TPOT, pueden aparecer diferentes modelos cada vez.
